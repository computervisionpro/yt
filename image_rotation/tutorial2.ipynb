# How to rotate an image using open cv ?

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt # importing packages
```

```python
# reading image

img_dir = 'C:\\Computer-Vision-Practice\\'
image = img_dir+'messi.jpg' # image path

im = cv2.imread(image)

plt.imshow(im)
```

```python
# above image is in bgr format...

im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

plt.imshow(im_rgb)
```

```python
# Now finally how to rotate?

h,w = im_rgb.shape[:2] # heught, width of image
center = (w//2,h//2)

M = cv2.getRotationMatrix2D(center, 10, .5)  # (center, angle of rotation, scale of zooming)

rotate=cv2.warpAffine(im_rgb, M, (w//2,h))
plt.imshow(rotate);
```
